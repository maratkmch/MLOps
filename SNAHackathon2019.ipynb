Конвертируем unixtime в timestamp и выделим часы

```python
df_hour = df[['instanceId_objectType', 'metadata_createdAt']]
```

```python
arr_post = df_hour[df_hour.instanceId_objectType=='Post'].sample(frac=0.01).metadata_createdAt
```

```python
arr_post_dt = pd.to_datetime(arr_post.head(10000).to_numpy(), unit='ms')
```

```python
from collections import Counter
```

```python
dict_post = dict(sorted(dict(Counter([t.hour for t in arr_post_dt])).items(), key=lambda item: item[0]))
```

```python
import matplotlib.pyplot as plt

plt.bar(list(dict_post.keys()), dict_post.values())
plt.xlabel("hour")
plt.ylabel("Count user")
plt.title("Post object")
```
